```python
import os
import pandas as pd
import numpy as np
import re
```

```python
path = '../Data/Preprocessed/탈모톡톡_탈모수다.xlsx'

df_origin = pd.read_excel(path)
df_origin['date'] = pd.to_datetime(df_origin['date'])
df_origin.loc[df_origin['rank'].isnull(), 'rank'] = 'any'

column_list = ['Fuck탈모', '고민상담', '나의치료법', '일반잡담', '주제토론', '탈모제품']

df_origin = df_origin[df_origin['category'].apply(lambda x : x in column_list)]
df_origin['keyword'] = df_origin['keyword'].apply(lambda x : eval(x))
```

```python
df_origin['keyword'] = df_origin['keyword'].apply(lambda x : eval(x))
```

```python

df_origin[(df_origin['propTopic'] == 6) & (df_origin['keyword'].apply(lambda x : '미녹시딜' in x ))].sort_values(by = ['view'], ascending = False)['content'].iloc[4]
```

```python
# 탈모수다: category 및 topic별 게시글 비율
df = df_origin.copy()
result = df.groupby(['category', 'propTopic'])['href'].count().reset_index()
result = result.pivot_table(index = ['propTopic'], columns = ['category'], values = ['href']).fillna(0)
for i in range(len(result.columns)):
    result.iloc[:, i] = round(result.iloc[:, i] / result.iloc[:, i].sum() * 100, 2)
result.T
```

```python
# 탈모수다: category 및 topic별 평균 조회수
df = df_origin.copy()
result = df.groupby(['category', 'propTopic'])['view'].mean().reset_index()
result = result.pivot_table(index = ['propTopic'], columns = ['category'], values = ['view']).fillna(0)
result = result.apply(lambda x : round(x))
result = result.astype(int)

result.T
```
