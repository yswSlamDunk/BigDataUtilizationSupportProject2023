```python
import os
import pandas as pd
import numpy as np
import re

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = '../Data/Preprocessed/탈모톡톡_19~23세탈모.xlsx'
```

```python
column_list = ['[22세탈모]', '[23세탈모]', '[20세탈모]', '[19세탈모]', '[21세탈모]', '[군대문제]']

df_origin = pd.read_excel(path)
df_origin['view'] = df_origin['view'].apply(lambda x : int(re.sub('[^0-9]', '', x)))
df_origin['date'] = pd.to_datetime(df_origin['date'])
df_origin['year'] = df_origin['date'].dt.year
df_origin['month'] = df_origin['date'].dt.month

condition00 = df_origin['date'] >= '2020'
condition01 = df_origin['category'].apply(lambda x : x in column_list)

df_origin = df_origin[condition01]
```

```python
df_origin.head(1)
```

```python
# rank별 작성 게시글 수
df = df_origin.copy()
df.groupby(['rank'])['title'].nunique().reset_index().to_excel('../Data/Result/19~23세탈모/rank별 게시글 수.xlsx', index = False)
    
```

```python
# rank별 게시글 평균 조회수
df = df_origin.copy()
df.groupby(['rank']).agg({'view' : ['mean', 'std']}).reset_index().to_excel('../Data/Result/19~23세탈모/rank별 평균 조회수.xlsx')
```

```python
# 월별 평균 조회수
df = df_origin.copy()
tmp = df.groupby(['month'])['view'].mean().reset_index()
tmp['view'] = tmp['view'].apply(lambda x : round(x, 2))
tmp['month'] = tmp['month'].apply(lambda x : str(x) + '월')
tmp
```

```python
# 월별 평균 게시글 수
df = df_origin.copy()
tmp = df.groupby(['year', 'month'])['href'].nunique().reset_index().groupby(['month'])['href'].mean().reset_index()
tmp['month'] = tmp['month'].apply(lambda x : str(x) + '월')
tmp['href'] = tmp['href'].apply(lambda x : round(x))
tmp
```

```python
# 월별 category별 평균 게시글 수
df = df_origin.copy()
tmp = df.groupby(['year', 'month', 'category'])['href'].nunique().reset_index().groupby(['month', 'category'])['href'].mean().reset_index()
tmp['month'] = tmp['month'].apply(lambda x : str(x) + '월')
tmp['href'] = tmp['href'].apply(lambda x : round(x))
tmp.pivot(columns = ['category'], index = ['month'], values = ['href'])
```

```python
# 월별 category별 평균 조회수
df = df_origin.copy()
tmp = df.groupby(['month', 'category'])['view'].mean().reset_index()
tmp['view'] = tmp['view'].apply(lambda x : round(x, 2))
tmp['month'] = tmp['month'].apply(lambda x : str(x) + '월')
tmp.pivot(columns = ['category'], index = ['month'], values = ['view'])
```
