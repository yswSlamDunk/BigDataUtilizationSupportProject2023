```python
import pandas as pd

```

```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('text_media1.csv', encoding='utf-8')

for col in df.columns:
    df[col] = df[col].replace(['(blank)'],'n/a')
    df[col] = df[col].replace(['(locked)'],'n/a')
df = df.loc[df['Title'] != 'n/a']

df['DateTime'] = pd.to_datetime(df['DateTime'],format='%Y-%m-%d %H:%M:%S')

np.histogram(df,100)
```
