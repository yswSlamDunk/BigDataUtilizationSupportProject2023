**`User Type` = 내부/외부 접속자인지 구분 (NA는 '기타'로 처리)**
- 내부: Referer에 'lafent.com'이나 Referer Search word에 'lafent'나 '라펜트'가 있을 때. 
- 외부: 'lafent'/'라펜트' 아닐 때
- 미확인: NaN

```python
# 'User Type'을 분류하는 함수 정의
def user_type(row):
    if pd.notna(row['Referer']) and 'lafent.com' in row['Referer']:
        return '내부'
    elif pd.notna(row['Referer Search Word']) and ('lafent' in row['Referer Search Word'] or '라펜트' in row['Referer Search Word']):
        return '내부'
    elif pd.isna(row['Referer']) and pd.isna(row['Referer Search Word']):
        return '미확인'
    else:
        return '외부'

# 'User Type' 칼럼 추가
data['User Type'] = data.apply(user_type, axis=1)
```

**`New IP` = 같은 날의 IP별로 IP주소 이름을 새로 설정**

```python
data["Date"] = data["Timestamp"].dt.strftime('%Y-%m-%d')
data['New IP'] = data.groupby(['Date', 'IP']).ngroup() + 1
#ngroup() = number each group from 0 to the number of groups -1 
```

**`New Category` = URL에서 'news_id:news', 'photo_id:photo' 추출하고, 'Category'에 news, photo 제외하고 값을 그대로 가져오기**

```python
#New Category를 계산하는 함수 정의 
def new_cat(row):
    if 'news_id' in row['URL']:
        return 'news'
    elif 'photo_id' in row['URL']:
        return 'photos'
    else:
        return row['Category']

# 'New Category' 칼럼 추가
data['New Category'] = data.apply(new_cat, axis=1)
```

**`Duration` = 같은 날의 IP별로 (마지막  Timestamp - 첫 Timestamp)**

```python
duration = data.copy()
duration['Duration'] = duration.groupby([duration['Timestamp'].dt.strftime('%Y-%m-%d'),'New IP'])['Timestamp'].transform(lambda x: x.max() - x.min())
duration = duration[['Timestamp', 'New IP', "Duration"]].drop_duplicates().reset_index(drop=True)
duration.head() #확인
```

Dataframe 3: Timestamp에서 날짜 기준 URL별 IP 수 및 고유한 IP 수  

```python
base0 = data.copy()
```

```python
#일별 URL별 총 IP 수 (페이지뷰) - not unique
base1 = base0.copy()
base1["date"] = base1['Timestamp'].dt.strftime('%Y-%m-%d')
pageview = base1.groupby(["date","URL"])["IP"].count().reset_index()
base1 = pd.DataFrame(pageview)
base1 = base1.rename(columns={'IP': 'pageview'})
base1.head()
```

```python
#일별 URL별 총 고유한 IP 수 - unique
base2 = base0.copy()
base2["date"] = base2['Timestamp'].dt.strftime('%Y-%m-%d')
customer = base2.groupby(['date', 'URL'])['IP'].nunique().reset_index()
base2 = pd.DataFrame(customer)
base2 = base2.rename(columns={'IP': 'customer'})
base2.head()
```

```python
print(base1.shape)
print(base2.shape)
```

```python
#base1 + base2 합치기 
base = pd.merge(base1, base2, on=["date", "URL"], how="inner")
base
```

```python
base.isna().sum()
```

```python
base.to_csv("0_base.csv", index=False)
```
