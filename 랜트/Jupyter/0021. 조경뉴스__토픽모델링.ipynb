```python
import numpy 
import pandas as pd
import numpy as np
import re
import pickle

from tqdm import tqdm
tqdm.pandas()

import gensim
from gensim import corpora

import matplotlib.pyplot as plt

import pyLDAvis
import pyLDAvis.gensim_models as pyGensim
import pickle

pyLDAvis.enable_notebook()
```

```python
path = '../Data/Preprocessed/조경뉴스_keywords.csv'
df_origin = pd.read_csv(path)
df_origin['keyword'] = df_origin['keyword'].apply(lambda x : eval(x))
df_origin['keywords'] = df_origin['keywords'].apply(lambda x : eval(x))
df_origin['keyword'] = df_origin['keyword'] + df_origin['keywords']
```

```python
df_origin.head(1)
```

```python
numTopics = [i for i in range(5, 21)]

def makeCorpus(series):
    # series = series.apply(lambda x : eval(x))
    dictionary = corpora.Dictionary(series)
    corpus = [dictionary.doc2bow(text) for text in series]
    return dictionary, corpus
```

```python
series = df_origin['keyword']
dictionary, corpus = makeCorpus(series)

for numTopic in tqdm(numTopics):
    ldaModel = gensim.models.ldamodel.LdaModel(corpus, num_topics = numTopic,
                                                id2word = dictionary, passes = 15)
    with open('../Result/LDA/' + f'topicNum_{str(numTopic)}.pickle', 'wb') as f:
        pickle.dump((ldaModel, dictionary, corpus), f)
```

```python
with open(file = '../Result/LDA/topicNum_5.pickle', mode = 'rb') as f:
    lda = pickle.load(f)
    ldaModel = lda[0]
    dictionary = lda[1]
    corpus = lda[2]

    vis = pyGensim.prepare(ldaModel, corpus, dictionary)

pyLDAvis.display(vis)
```
