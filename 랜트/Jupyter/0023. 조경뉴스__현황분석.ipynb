```python
import os
import pandas as pd
import numpy as np

from tqdm import tqdm
tqdm.pandas()
```

```python
path = '../Data/Preprocessed/topicResult_10_summary_tags.csv'

df_origin = pd.read_csv(path, encoding = 'cp949')
df_origin['date'] = pd.to_datetime(df_origin['date'])
df_origin.set_index('date', inplace = True)
```

```python
innerOut_df_origin = pd.read_csv('../Data/Preprocessed/innerOutTaged.csv')
innerOut_df_origin['Timestamp'] = pd.to_datetime(innerOut_df_origin['Timestamp'])
innerOut_df_origin.set_index('Timestamp', inplace = True)
```

```python
df_origin.head(1)
```

```python
# publicher별 게시글 비율
result = df_origin.groupby(['publisher'])['href'].count().reset_index()
print('총 데이터 수:', len(df_origin))
result
```

```python
# 게시글 작성 추이
df = df_origin.copy()
df = df.resample('1M')['href'].count().reset_index()
df['date'] = df['date'].astype(str)
df['date'] = df['date'].str[:7]
```

```python
print('전체 기간의 월별 평균 조회수:', df.mean())
print('2021년 05월 이전의 월별 평균 조회수:', df.loc[df['date'] < '2021-05'].mean())
print('2021년 05월 이후의 월별 평균 조회수:', df.loc[df['date'] >= '2021-05'].mean())
```

```python
# 게시글 수와 페이지뷰 수 관계 분석
df = df_origin.copy()
df = df.loc['2022/08/01':]
print('2022/08/01 이후에 작성된 기사 수:', len(df))
innerOut_df = innerOut_df_origin.copy()
```

```python
userType_pageView = innerOut_df.groupby(['User Type']).resample('1M')['URL'].count().reset_index()
userType_pageView = userType_pageView.pivot_table(index = ['Timestamp'], columns = ['User Type'], values = ['URL']).fillna(0).reset_index()
userType_pageView[('URL', '전체')] = userType_pageView[[('URL', '내부'), ('URL', '외부')]].sum(axis = 1)
userType_pageView
```

```python
userType_newsId_pageView = innerOut_df.loc[(innerOut_df['URL'].str.contains('news_id'))].groupby(['User Type']).resample('1M')['URL'].count().reset_index()
userType_newsId_pageView = userType_newsId_pageView.pivot_table(index = ['Timestamp'], columns = ['User Type'], values = ['URL']).fillna(0)
userType_newsId_pageView[('URL', '전체')] = userType_newsId_pageView[[('URL', '내부'), ('URL', '외부')]].sum(axis = 1)
userType_newsId_pageView
```

```python
# 게시글 수와 방문객 관계 분석
userType_newIp = innerOut_df.groupby(['User Type']).resample('1M')['New IP'].nunique().reset_index()
userType_newIp = userType_newIp.pivot_table(index = ['Timestamp'], columns = ['User Type'], values = ['New IP']).fillna(0).reset_index()
userType_newIp[('New IP', '전체')] = userType_newIp[[('New IP', '내부'), ('New IP', '외부')]].sum(axis = 1)
userType_newIp
```

```python
userType_newIp = innerOut_df.loc[(innerOut_df['URL'].str.contains('news_id'))].groupby(['User Type']).resample('1M')['New IP'].nunique().reset_index()
userType_newIp = userType_newIp.pivot_table(index = ['Timestamp'], columns = ['User Type'], values = ['New IP']).fillna(0).reset_index()
userType_newIp[('New IP', '전체')] = userType_newIp[[('New IP', '내부'), ('New IP', '외부')]].sum(axis = 1)
userType_newIp
```
