```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm

tqdm.pandas()
```

```python
dir_path = '../Data/Raw/Revenue'

df_origin = pd.DataFrame()

for path in os.listdir(dir_path):
    tmp = pd.read_csv(os.path.join(dir_path, path))
    df_origin = pd.concat([df_origin, tmp], axis = 0)

df_origin = df_origin[~pd.isnull(df_origin['국가'])]    
df_origin['HS코드'] = df_origin['HS코드'].astype(int).astype(str)
df_origin = df_origin.loc[((df_origin['국가'] == '중국') & (df_origin['HS코드'] == '903040')) | ((df_origin['국가'] != '중국') & (df_origin['HS코드'] == '903090'))]
df_origin['년'] = df_origin['기간'].apply(lambda x : x.split('-')[0])
df_origin['월'] = df_origin['기간'].apply(lambda x : x.split('-')[1])

# df_origin.to_csv('../Data/Preprocessed/revenue.csv', index = False, encoding = 'utf-8-sig', escapechar = '\\')
```
